2) Seleccionar las columnas de interés. ¿Qué hago con los NaN?

Comenzamos leyendo una muestra de 10000 registros del fichero "bookings", seleccionando las columnas de interés (aeropuerto, número de pasajeros y año):

```python
size_sm = 10000
b = pd.read_csv('./bookings.csv.bz2', sep = "^", usecols = ['arr_port','pax','year'], 
                nrows = size_sm)

#Eliminamos los NaN para no arrastraslos
b = b.dropna()

b.sample(5)
```

3) Plan de acción.

Tenemos que sacar el top 10 de aeropuertos de llegada en el año 2013 y el número de pasajeros que han aterrizado en cada aeropuerto. Entonces, filtraremos por el año, agruparemos por aeropuertos de llegada y sumaremos el número de pasajeros:

```python
#Filtramos por el año
b = b[b["year"] == 2013]

#Agrupamos aeropuertos y sumamos pasajeros
b_gr = b.groupby(['arr_port']).sum().sort_values('pax',ascending = False)

#Como la columna 'year' no es necesaria, la eliminamos
del b_gr['year']

#Para hacer mejor el DataFrame, reestablecemos su índice
print('Top 10 arrival airports in the world in 2013:')
b_gr.reset_index().head(10)
```

4) Ajustar el código para trabajar con Big Data

Por último, ajustamos el código anterior para trabajar con Big Data aplicando los Chunks:

```python
#Leemos el fichero "bookings" fijando el número de filas que importaremos al DF en cada iteración con el tamaño de los Chunks
size_sm = 10000
size_chk = 3500
bk = pd.read_csv('./bookings.csv.bz2', sep = "^", usecols = ['arr_port','pax','year'], 
                 nrows = size_sm, chunksize = size_chk) 
#Creamos un DF vacío donde vamos a ir añadiendo los resultados de los Chunks
bk_chunks = pd.DataFrame()

#Para cada iteración de tamaño definido en el 'chunksize', filtraremos por el año, agruparemos por aeropuerto, sumaremos los pasajeros,
#y lo iremos añadiendo al DF vacío.
for i, b in enumerate(bk):
    b = b.dropna() #eliminamos los NaN
    b = b[b["year"] == 2013]
    del b['year'] 
    b_gr = b.groupby(['arr_port']).sum().sort_values('pax',ascending = False) 
    b_gr.reset_index(inplace = True)
    bk_chunks = bk_chunks.append(b_gr)

#Volvemos a agrupar por aeropuerto, sumamos los pasajeros y restablecemos el índice. Pero esta vez, para el DF obtenido de las iteraciones.
bk_gr = bk_chunks.groupby(['arr_port']).sum().sort_values('pax',ascending = False)
bk_gr.reset_index(inplace = True)
print('Top 10 arrival airports in the world in 2013:')
bk_gr.head(10)
```
