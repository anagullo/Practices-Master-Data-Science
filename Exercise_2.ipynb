2) Seleccionar las columnas de inter√©s. ¬øQu√© hago con los NaN?

Comenzamos leyendo una muestra de 10000 registros del fichero "bookings", seleccionando las columnas de inter√©s (aeropuerto, n√∫mero de pasajeros y a√±o):

```python
size_sm = 10000
b = pd.read_csv('./bookings.csv.bz2', sep = "^", usecols = ['arr_port','pax','year'], 
                nrows = size_sm)

#Eliminamos los NaN para no arrastraslos
b = b.dropna()

b.sample(5)
```

3) Plan de acci√≥n.

Tenemos que sacar el top 10 de aeropuertos de llegada en el a√±o 2013 y el n√∫mero de pasajeros que han aterrizado en cada aeropuerto. Entonces, filtraremos por el a√±o, agruparemos por aeropuertos de llegada y sumaremos el n√∫mero de pasajeros:

```python
#Filtramos por el a√±o
b = b[b["year"] == 2013]

#Agrupamos aeropuertos y sumamos pasajeros
b_gr = b.groupby(['arr_port']).sum().sort_values('pax',ascending = False)

#Como la columna 'year' no es necesaria, la eliminamos
del b_gr['year']

#Para hacer mejor el DataFrame, reestablecemos su √≠ndice
print('Top 10 arrival airports in the world in 2013:')
b_gr.reset_index().head(10)
```

4) Ajustar el c√≥digo para trabajar con Big Data

Por √∫ltimo, ajustamos el c√≥digo anterior para trabajar con Big Data aplicando los Chunks:

```python
#Leemos el fichero "bookings" fijando el n√∫mero de filas que importaremos al DF en cada iteraci√≥n con el tama√±o de los Chunks
size_sm = 10000
size_chk = 3500
bk = pd.read_csv('./bookings.csv.bz2', sep = "^", usecols = ['arr_port','pax','year'], 
                 nrows = size_sm, chunksize = size_chk) 
#Creamos un DF vac√≠o donde vamos a ir a√±adiendo los resultados de los Chunks
bk_chunks = pd.DataFrame()

#Para cada iteraci√≥n de tama√±o definido en el 'chunksize', filtraremos por el a√±o, agruparemos por aeropuerto, sumaremos los pasajeros,
#y lo iremos a√±adiendo al DF vac√≠o.
for i, b in enumerate(bk):
    b = b.dropna() #eliminamos los NaN
    b = b[b["year"] == 2013]
    del b['year'] 
    b_gr = b.groupby(['arr_port']).sum().sort_values('pax',ascending = False) 
    b_gr.reset_index(inplace = True)
    bk_chunks = bk_chunks.append(b_gr)

#Volvemos a agrupar por aeropuerto, sumamos los pasajeros y restablecemos el √≠ndice. Pero esta vez, para el DF obtenido de las iteraciones.
bk_gr = bk_chunks.groupby(['arr_port']).sum().sort_values('pax',ascending = False)
bk_gr.reset_index(inplace = True)
print('Top 10 arrival airports in the world in 2013:')
bk_gr.head(10)
```
